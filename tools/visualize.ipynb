```python
from matplotlib import pyplot as plt
def visualize(image, cmap="viridis"):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(image, cmap=cmap)
```

```python
from optic.data import FundusDataset

data_root="../data/GAMMA_training_data/training_data"

dataset = FundusDataset(
    data_root=data_root,
    mode="trainval",
    return_oct=True
)
```

```python
from optic.data import img_augment, oct_augment

img_aug = img_augment(is_train=True)

sample = dataset[0]
img = sample["img"]
print(sample["id"])
visualize(img)
aug_img = img_aug(image=img)["image"]
visualize(aug_img)
```

```python
oct_img = sample["oct_img"]
oct_aug = oct_augment()
print(oct_img.shape)
visualize(oct_img[:,:,100], "gray")
visualize(oct_img[:,:,200], "gray")
auged_oct_img = oct_aug(image=oct_img)["image"]
print(auged_oct_img.shape)
visualize(auged_oct_img[:,:,100], "gray")
visualize(auged_oct_img[:,:,200], "gray")
```
